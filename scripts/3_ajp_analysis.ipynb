```python
lem_sents_all = []
for filename in [f for f in os.listdir("../data/lemmatized_sents") if ".txt" in f]:
    id = filename.partition(".")[0]
    with open("../data/lemmatized_sents/" + filename, "r") as f:
        lines = f.readlines()
    lines = [{"id": id, "sent" : line.replace("\n", "")} for line in lines]
    lem_sents_all.extend(lines)
```
