```python
import stanza
import os
```

```python
nlp = stanza.Pipeline('cs')
```

```python
with open("../data/raw_texts/2651_1974.txt", "r") as f:
    text = f.read()
```

```python
text
```

```python
doc = nlp(text)
```

```python
tags = ["NOUN", "ADJ", "PROPN", "VERB"]
```

```python
lemmatized_sents = []
for sentence in doc.sentences:
    lemmatized_sents.append(" ".join([word.lemma for word in sentence.words if word.pos in tags]))
```

```python
for filename in [f for f in os.listdir("../data/raw_texts") if ".txt" in f]:
    with open("../data/raw_texts/" + filename, "r") as f:
        text = f.read()
    if len(text) > 0:
        doc = nlp(text)
        lemmatized_sents = []
        for sentence in doc.sentences:
            lemmatized_sents.append(" ".join([word.lemma for word in sentence.words if word.pos in tags]) + "\n")
        filename_new = filename.partition("_")[0] + ".txt"
        with open("../data/lemmatized_sents/" + filename_new, "w") as f:
            f.writelines(lemmatized_sents)
```
